```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=data.drop('Species',axis=1)
```

```python
y=data['Species']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log =LogisticRegression()
log.fit(X_train,Y_train)
prediction=log.predict(X_test)
from sklearn import metrics
print( metrics.accuracy_score(prediction,Y_test))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
pre=knn.predict(X_test)
print(metrics.accuracy_score(pre,Y_test))
```

```python

#SupportVectorMAchine
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
pred=svc.predict(X_test)
print(metrics.accuracy_score(pred,Y_test))
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
prew=dtc.predict(X_test)
print(metrics.accuracy_score(prew,Y_test))
```

```python
#RAndomForest
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
p=rfc.predict(X_test)
print(metrics.accuracy_score(p,Y_test))
```

```python
#NAiveBayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
pr=nb.predict(X_test)
print(metrics.accuracy_score(pr,Y_test))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,pr))
```
